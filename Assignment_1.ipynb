```python
import cv2
import numpy as np
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

image_url = "https://images5.alphacoders.com/830/830408.jpg"

#  download and read image from URL
def load_image_from_url(image_url):
    response = requests.get(image_url)
    image = Image.open(BytesIO(response.content))
    return np.array(image)

# extract a 100Ã—100 center region
def extract_center_region(image):
    h, w = image.shape[:2]
    center_x, center_y = w // 2, h // 2
    x1, y1 = center_x - 50, center_y - 50
    x2, y2 = center_x + 50, center_y + 50
    return image[y1:y2, x1:x2]

# Load image
image = load_image_from_url(image_url)

# Display the full image before cropping
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()

# Extract center crop
center_crop = extract_center_region(image)

# Display extracted 100x100 region
plt.figure(figsize=(3, 3))
plt.imshow(center_crop)
plt.axis("off")
plt.title("Center Crop (100x100)")
plt.show()
```

```python
import cv2
import numpy as np
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

image_url = "https://images5.alphacoders.com/830/830408.jpg"

#  download and read image from URL
def load_image_from_url(image_url):
    response = requests.get(image_url)
    image = Image.open(BytesIO(response.content))
    return np.array(image)

#  convert to HSV and visualize components
def convert_to_hsv_and_visualize(image):
    # Convert image to BGR if it's in RGB
    if len(image.shape) == 3 and image.shape[2] == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv_image)

    # e HSV components in grayscale
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(h, cmap="gray")
    axes[0].set_title("Hue Component")
    axes[0].axis("off")

    axes[1].imshow(s, cmap="gray")
    axes[1].set_title("Saturation Component")
    axes[1].axis("off")

    axes[2].imshow(v, cmap="gray")
    axes[2].set_title("Value Component")
    axes[2].axis("off")

    plt.show()

# Load image
image = load_image_from_url(image_url)

# Converting to HSV and visualize component
convert_to_hsv_and_visualize(image)
```
