```python
#Assignment 03
#Arnob Bokshi
#2125051010
import cv2
import numpy as np
import matplotlib.pyplot as plt

# two 400x400 black images
image1 = np.zeros((400, 400, 3), dtype=np.uint8)
image2 = np.zeros((400, 400, 3), dtype=np.uint8)

# a green rectangle at the center of the first image
cv2.rectangle(image1, (100, 150), (300, 250), (0, 255, 0), -1)  # Green (BGR format)

#  magenta circle at the center of the second image
cv2.circle(image2, (200, 200), 75, (255, 0, 255), -1)  # Magenta (BGR format)

# Function to display images using Matplotlib
def show_image(image, title):
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()

# display original images
show_image(image1, "Image 1 (Green Rectangle)")
show_image(image2, "Image 2 (Magenta Circle)")

```

```python

#  bitwise AND
bitwise_and = cv2.bitwise_and(image1, image2)
show_image(bitwise_and, "Bitwise AND")
```

```python

#  bitwise OR
bitwise_or = cv2.bitwise_or(image1, image2)
show_image(bitwise_or, "Bitwise OR")
```

```python

# bitwise XOR
bitwise_xor = cv2.bitwise_xor(image1, image2)
show_image(bitwise_xor, "Bitwise XOR")
```

```python

#  bitwise NOT on Image 1
bitwise_not1 = cv2.bitwise_not(image1)
show_image(bitwise_not1, "Bitwise NOT (Image 1)")
```

```python

# bitwise NOT on Image 2
bitwise_not2 = cv2.bitwise_not(image2)
show_image(bitwise_not2, "Bitwise NOT (Image 2)")
```

```python

#  element-wise multiplication
multiplication = cv2.multiply(image1, image2)
show_image(multiplication, "Multiplication")
```

```python

#  alpha blending
alpha = 0.5
blended = cv2.addWeighted(image1, alpha, image2, 1 - alpha, 0)
show_image(blended, "Alpha Blending")
```

```python

#  image subtraction
subtraction = cv2.subtract(image1, image2)
show_image(subtraction, "Subtraction")
```
