```python
import requests
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
url = "https://images.pexels.com/photos/29724751/pexels-photo-29724751.jpeg?auto=compress&cs=tinysrgb&w=600&lazy=load"
response = requests.get(url)
image = Image.open(BytesIO(response.content))
image = np.array(image)

#original image
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
# task a . Resize to 256x256
resized_image = cv2.resize(image, (256, 256))
plt.imshow(resized_image)
plt.axis("off")
plt.title("Resized 256x256")
plt.show()
```

```python
# task b. Rotate by 180 degrees clockwise
rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)
plt.imshow(rotated_image)
plt.axis("off")
plt.title("Rotated 180°")
plt.show()
```

```python
# task c. vertical flip
flipped_image = cv2.flip(resized_image, 0)
plt.imshow(flipped_image)
plt.axis("off")
plt.title("Vertically Flipped")
plt.show()
```

```python
# task d. Displayed all transformations in a subplot
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(resized_image)
axes[0].set_title("Resized 256x256")
axes[0].axis("off")
axes[1].imshow(rotated_image)
axes[1].set_title("Rotated 180°")
axes[1].axis("off")
axes[2].imshow(flipped_image)
axes[2].set_title("Vertically Flipped")
axes[2].axis("off")
plt.tight_layout()
plt.show()
```
